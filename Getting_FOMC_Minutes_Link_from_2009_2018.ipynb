```
### getting all minutes in htm 


from __future__ import print_function
import requests
from bs4 import BeautifulSoup, Comment
import pandas as pd
import re
import matplotlib.pyplot as plt

# Input year range 
range_of_years = range(2009,2019)


# Part one => For FOMC before 2013, minutes are in individual pages : https://www.federalreserve.gov/monetarypolicy/fomchistorical{year}.htm

# Creating the link for each year 

web_front = 'https://www.federalreserve.gov/monetarypolicy/fomchistorical'
web_end = '.htm'

main_url_each_year = [] 
for year in range_of_years :
  
  each_url = web_front + str(year) + web_end
  main_url_each_year.append(each_url)
  
links1 = []

# Getting links from each page and search for fomc minutes htm links 

for links_each_year in main_url_each_year:
  
 
  content = requests.get(links_each_year).text
  soup = BeautifulSoup(content, 'html')


  for a in soup.find_all('a', href=True):
      links1.append(str(a['href']))
  
  links1 = [s for s in links1 if  "minutes" in s]

  links1 = [s for s in links1 if ".htm" in s]
  conf_call_min = [ s for s in links1 if '#' in s ]
  links1 = [s for s in links1 if "#" not in s]
  
# Getting the full link 

  web = 'https://www.federalreserve.gov'

  
  links2 = []
  for s in links1:

    if 'http' not in s :
      links2.append(web +s)
    else: 
      links2.append(s)

# all minutes from 2013 to 2018 

from __future__ import print_function
url='https://www.federalreserve.gov/monetarypolicy/fomccalendars.htm'
content = requests.get(url).text
soup = BeautifulSoup(content, 'html')

links = []
for a in soup.find_all('a', href=True):
    links.append(str(a['href']))

links = [s for s in links if "fomcminutes" in s]
links = [s for s in links if ".htm" in s]

web = 'https://www.federalreserve.gov'
links = [web + s  for s in links]

link_total = links + links2 


# Printing out the link 

print(len(link_total))

year_range = range_of_years

for year in year_range:
  links1 = [s for s in link_total if str(year) in s]
  print('Year:' , year)
  print(*links1, sep='\n')
  print(len(links1))
  print('\n')
```
